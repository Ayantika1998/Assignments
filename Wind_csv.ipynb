```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(url, sep = "\s+", parse_dates = [[0,1,2]]) #replacing the first 3 columns by a proper datetime index
data.head()
```

```python
data.info()
```

**Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.**

```python
#creating a function to fix the date
# Creating a function to correct the years
def years(y):
  year = y.year - 100 if y.year > 1989 else y.year
  return datetime.date(year, y.month, y.day)

# applying the function
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(years)

# Looking at the data now
data.head()
```

**Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].**

```python

data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy') # setting the correct date column 'Yr_Mo_Dy'
data.head()
```

**Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
A different set of numbers for each day.**

```python
# Statistics for each day at all locations
day_stats = pd.DataFrame()

# All statistics:
day_stats['min'] = dataset.min(axis = 1) # returning minimum
day_stats['max'] = dataset.max(axis = 1) # returning maximum 
day_stats['mean'] = dataset.mean(axis = 1) # returning mean
day_stats['std'] = dataset.std(axis = 1) # returning standard deviations

day_stats.head()
```

**Step 12. Downsample the record to a yearly frequency for each location.**

**Step 13. Downsample the record to a monthly frequency for each location.**

**Step 14. Downsample the record to a weekly frequency for each location.**
