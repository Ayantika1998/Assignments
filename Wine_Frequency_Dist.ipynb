```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
wine.keys()
```

```python
print(wine['DESCR'])
```

```python
print(wine["target_names"])
```

```python
df=pd.DataFrame(wine["data"],columns=wine["feature_names"])
```

```python
df.head()
```

```python
df["class"]=wine["target"]
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
phenols_freq=pd.crosstab(index=df["nonflavanoid_phenols"], columns='Frequency')
phenols_freq
```

```python
sns.histplot(phenols_freq,kde=False, alpha=0.5)
```

```python
alcohol_freq=pd.crosstab(df["alcohol"],columns='Frequency')
alcohol_freq
```

```python
sns.histplot(alcohol_freq,kde=False, alpha=0.5)
```

```python
sns.histplot(malic_acid_freq,kde=False, alpha=0.5)
```
