### **Dummification**

A One hot encoding is a representation of categorical variable as binary vectors.It allows the representation of categorical data to be more expresive. This first requires that the categorical values be mapped to integer values, that is label encoding. Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.

The Dummy variable trap is a scenario in which the independent variable are multicollinear, a scenario in which two or more variables are highly correlated in simple term one variable can be predicted from the others.

By using pandas get_dummies function we can do all above three step in line of code. We will this fuction to get dummy variable for sex, children,smoker,region features. By setting drop_first =True function will remove dummy variable trap by droping one variable and original variable.

# **Model Evaluation**
# **Decision Tree vs Random Forest**

**R-Squared (R)**
R represents the proportion of variance explained by your model.

R is a relative metric, so you can use it to compare with other models trained on the same data. And you can use it to get a rough a feel for how well a model performs, in general.
A model that explains no variance would have an R of  A model with an R of 1 would explain all of the variance. Higher scores are better.

**Root Mean Squared Error (RMSE)**
RMSE is a very common evaluation metric. It can range between 0 and infinity. Lower values are better. 

**Mean Absolute Error**
Mean Absolute Error (MAE) is the average of the absolute value of the errors.

**MSE**
The Mean Squared Error measures how close a regression line is to a set of data points. It is a risk function corresponding to the expected value of the squared error loss.

### **Decision Tree Regression**

```python
score = r2_score(Y_test,predict)
print("r_square score --> ",score)
print('Mean Absolute Error -->', metrics.mean_absolute_error(Y_test, predict))
print('Mean Squared Error -->', metrics.mean_squared_error(Y_test, predict))
print('Root Mean Squared Error -->', np.sqrt(metrics.mean_squared_error(Y_test, predict)))
```

### **Random Forest Regression**

```python
score_2 = r2_score(Y_test,predict_F)
print("r_square score --> ",score_2)
print('Mean Absolute Error -->', metrics.mean_absolute_error(Y_test, predict_F))
print('Mean Squared Error -->', metrics.mean_squared_error(Y_test, predict_F))
print('Root Mean Squared Error -->', np.sqrt(metrics.mean_squared_error(Y_test, predict_F)))
```

**It is clear from the above metrics that Random Forest Regression Model is somewhat better fit for this data than Decision Tree Regrssion Model. This is because we know a random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.**
